```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv' , encoding='unicode_escape')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull()  # or pd.isnull(df)
```

```python
df.isnull().sum() # or pd.isnull(df).sum()
# executed again after executong df.dropna() command so no o/p value for amount section.
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
df.rename(columns = {'Marital_Status':'Shaadi'})
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax =sns.countplot(x='Gender',data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

here we can see money is spend more by women.

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
sns.countplot(data = df, x= 'Age Group', hue='Gender')
```

# AGE

```python
ax = sns.countplot(data = df, x= 'Age Group', hue='Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

This indicates in age group 26-35 women are spending more

```python
sales_age = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data = sales_age)
```

From this we can see that most of the buyers are of group between 26-35 yrs female

# STATE

```python
sales_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x='State',y='Amount')
```

this graph shows that most of the buyers are from uttar pradesh maharashtra karnatak. And highest are from uttar pradesh as shown

```python
sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x='State',y='Orders')
```

by these two graph we can conclude that even after placing less amount of orders haryana has spent more compared to kerala. As karala has placed more orders compared to haryana.

```python

sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by ='Amount', ascending=False)

sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x='Occupation',y='Amount')
```

this graph shows that peope from IT sector have higher purchasing power compared to others, as this graph shows the amount of money spend by different sectors.

# PRODUCT CATEGORY

```python
sns.set(rc= {'figure.figsize' :(24,7)})

ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

from above graph it is clear that most sold products are from clothing food and electronics.

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x='Product_Category',y='Amount')
```

If we comapare the amount of  money spend on specific categories compared to the total product sold , we can see that people have spend a lot more on food sector.

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x ='Product_ID',y='Orders')
```
